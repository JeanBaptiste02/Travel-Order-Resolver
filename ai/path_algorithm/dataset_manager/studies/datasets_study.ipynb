```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le fichier CSV
file_path = r"C:\Users\vikne\Documents\Master 2\Semestre 9\Intelligence artificielle\Travel-Order-Resolver\ai\path_algorithm\dataset\timetables.csv"

# Charger le dataset en DataFrame
df = pd.read_csv(file_path, sep='\t')  # Utiliser '\t' comme s√©parateur si le fichier est tabul√©

# Afficher les 5 premi√®res lignes
print("Les 5 premi√®res lignes du dataset :")
display(df.head())

# Graphique 1 : R√©partition des trajets les plus fr√©quents
plt.figure(figsize=(10, 6))
most_common_trips = df['trajet'].value_counts().head(10)
most_common_trips.plot(kind='bar', color='skyblue')
plt.title('Top 10 des trajets les plus fr√©quents')
plt.xlabel('Trajet')
plt.ylabel('Nombre de trajets')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Interpr√©tation des donn√©es des trajets les plus fr√©quents
print("\nInterpr√©tation des trajets les plus fr√©quents :")
print("Les 10 trajets les plus fr√©quents sont :")
print(most_common_trips)

# Graphique 2 : Distribution des dur√©es des trajets
plt.figure(figsize=(10, 6))
sns.histplot(df['duree'], bins=30, kde=True, color='orange')
plt.title('Distribution des dur√©es des trajets')
plt.xlabel('Dur√©e (minutes)')
plt.ylabel('Fr√©quence')
plt.tight_layout()
plt.show()

# Interpr√©tation de la distribution des dur√©es des trajets
print("\nInterpr√©tation de la distribution des dur√©es des trajets :")
duration_mean = df['duree'].mean()
duration_median = df['duree'].median()
duration_std = df['duree'].std()
duration_min = df['duree'].min()
duration_max = df['duree'].max()

print(f"Dur√©e moyenne des trajets : {duration_mean:.2f} minutes")
print(f"Dur√©e m√©diane des trajets : {duration_median:.2f} minutes")
print(f"√âcart-type des dur√©es : {duration_std:.2f} minutes")
print(f"Dur√©e minimale des trajets : {duration_min} minutes")
print(f"Dur√©e maximale des trajets : {duration_max} minutes")

# Conclusion rapide
print("\nLe dataset a √©t√© charg√© et analys√© avec des graphiques.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le fichier CSV
file_path = r"C:\Users\vikne\Documents\Master 2\Semestre 9\Intelligence artificielle\Travel-Order-Resolver\ai\path_algorithm\dataset\timetables_with_cities.csv"

# Charger le dataset en DataFrame
df = pd.read_csv(file_path, sep=';')
print("Les 5 premi√®res lignes du dataset :")
display(df.head())

# 1. Graphique 1 : Distribution des dur√©es des trajets (avec √©chelle logarithmique)
plt.figure(figsize=(10, 6))
sns.histplot(df['duree'], bins=50, kde=True, color='lightblue', log_scale=(False, True))
plt.title('Distribution des dur√©es des trajets (√©chelle log pour la dur√©e)')
plt.xlabel('Dur√©e des trajets (minutes)')
plt.ylabel('Fr√©quence')
plt.tight_layout()
plt.show()

# Interpr√©tation du graphique 1
print("\nInterpr√©tation de la distribution des dur√©es des trajets (√©chelle log) :")
min_duree = df['duree'].min()
max_duree = df['duree'].max()
mean_duree = df['duree'].mean()
median_duree = df['duree'].median()

print(f"Dur√©e minimale des trajets : {min_duree} minutes")
print(f"Dur√©e maximale des trajets : {max_duree} minutes")
print(f"Dur√©e moyenne des trajets : {mean_duree:.2f} minutes")
print(f"Dur√©e m√©diane des trajets : {median_duree:.2f} minutes")

# 2. Graphique 2 : Densit√© des dur√©es des trajets par groupe de trajets
plt.figure(figsize=(10, 6))
sns.kdeplot(df['duree'], fill=True, color='lightgreen', label='Dur√©e des trajets')
plt.title('Densit√© des dur√©es des trajets')
plt.xlabel('Dur√©e des trajets (minutes)')
plt.ylabel('Densit√©')
plt.tight_layout()
plt.show()

# Interpr√©tation du graphique 2
print("\nInterpr√©tation de la densit√© des dur√©es des trajets :")
density_mean_duree = df['duree'].mean()
density_median_duree = df['duree'].median()

print(f"La densit√© des trajets est concentr√©e autour de la dur√©e moyenne de {density_mean_duree:.2f} minutes,")
print(f"et la m√©diane se situe √† {density_median_duree:.2f} minutes.")

# 3. Graphique 3 : Dur√©e des trajets par quantiles
quantiles = [0.25, 0.5, 0.75]
df['quantile'] = pd.qcut(df['duree'], q=[0] + quantiles + [1], labels=['Q1', 'Q2', 'Q3', 'Q4'])

plt.figure(figsize=(10, 6))
sns.boxplot(x='quantile', y='duree', data=df, palette='coolwarm')
plt.title('Dur√©e des trajets par quantiles')
plt.xlabel('Quantiles de dur√©e')
plt.ylabel('Dur√©e des trajets (minutes)')
plt.tight_layout()
plt.show()

# Interpr√©tation du graphique 3
print("\nInterpr√©tation de la dur√©e des trajets par quantiles :")
q1 = df[df['quantile'] == 'Q1']['duree'].mean()
q2 = df[df['quantile'] == 'Q2']['duree'].mean()
q3 = df[df['quantile'] == 'Q3']['duree'].mean()
q4 = df[df['quantile'] == 'Q4']['duree'].mean()

print(f"Dur√©e moyenne du premier quantile (Q1) : {q1:.2f} minutes")
print(f"Dur√©e moyenne du deuxi√®me quantile (Q2) : {q2:.2f} minutes")
print(f"Dur√©e moyenne du troisi√®me quantile (Q3) : {q3:.2f} minutes")
print(f"Dur√©e moyenne du quatri√®me quantile (Q4) : {q4:.2f} minutes")

# 4. Graphique 4 : Heatmap de la fr√©quence des trajets par dur√©e et quantile
pivot_table = df.pivot_table(values='trajet', index='quantile', columns='duree', aggfunc='count', fill_value=0)

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', cbar_kws={'label': 'Fr√©quence des trajets'}, linewidths=0.5)
plt.title('Fr√©quence des trajets par dur√©e et quantile')
plt.xlabel('Dur√©e des trajets (minutes)')
plt.ylabel('Quantile de dur√©e')
plt.tight_layout()
plt.show()

# Interpr√©tation du graphique 4
print("\nInterpr√©tation de la heatmap de la fr√©quence des trajets :")
max_frequency = pivot_table.max().max()
max_freq_duration = pivot_table.stack().idxmax()

print(f"La fr√©quence maximale des trajets est de {max_frequency} trajets pour la dur√©e {max_freq_duration[1]} minutes dans le quantile {max_freq_duration[0]}.")
```
